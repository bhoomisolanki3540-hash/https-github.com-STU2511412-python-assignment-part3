```python
### File I/O, APIs & Exception Handling
##Theme: Product Explorer & Error-Resilient Logger
#Task 1 — File Read & Write Basics
file_name = "python_notes.txt"

#-------Part A : Write & Append--------

#1. Write the initial 5 topics
initial_notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]
try:
    with open(file_name, 'w', encoding="utf-8") as file:
        for line in initial_notes:
            file.write(line + "\n")
    print("File written successfully with initial notes.")

    # Append 2 more lines
    extra_notes = [
        "Topic 6: Functions promote code reusability.",
        "Topic 7: Modules allow organizing code into separate files."
    ]
    with open(file_name, 'a', encoding="utf-8") as file:
        for line in extra_notes:
            file.write(line + "\n")
    print("Extra notes appended successfully.")

except Exception as e:
    print(f"An error occurred while writing to the file: {e}")    

#-------Part B : Read & Search--------
print("\n----Reading the Python Notes----")
try:
    with open(file_name, 'r', encoding="utf-8") as file:
        lines = file.readlines()

    #Print numbered lines (stripping \n)    
    print("File read successfully. Content:")
    for index,line in enumerate (lines, start=1):
        print(f"{index}. {line.strip()}")

    # Count total lines
    print(f"\nTotal lines in the file: {len(lines)}")

    # Keyword search
    search_query = input("\nEnter a keyword to search for: ").strip().lower()
    matches = []  
    print(f"Searching for matches containing '{search_query}'...")
    
    for line in lines:
        # Case sensitivity check
        if search_query in line.lower():
            matches.append(line.strip())

    if matches:
        print(f"Found {len(matches)} matches for the keyword '{search_query}':")
        for match in matches:
            print(f"-> {match}")
    else:
        print(f"No matches found for the keyword '{search_query}'.")

except FileNotFoundError:
    print(f"Error: The file '{file_name}' does not exist. Please run the Write script first to create the file.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```

```python
##Task 2 — API Integration 
import requests


#Step 1 — Fetch and Display Products:
base_url = "https://dummyjson.com/products"
response = requests.get(f"{base_url}?limit=20")  # Fetch first 20 products

if response.status_code == 200:
    data = response.json()
    products = data["products"]

    print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<10} | {'Rating'}")
    print("-" * 75)
    for p in products:
        print(f"{p['id']:<4} | {p['title']:<30} | {p['category']:<15} | ${p['price']:<7.2f} | {p['rating']}")

    # --- Step 2: Filter (Rating >= 4.5) and Sort (Price Descending) ---
    print("\n--- High-Rated Products (Sorted by Price Desc) ---")
    
    # Filter logic
    high_rated = [p for p in products if p['rating'] >= 4.5]
    
    # Sorting logic (Price descending)
    sorted_products = sorted(high_rated, key=lambda x: x['price'], reverse=True)

    for p in sorted_products:
        print(f"Price: ${p['price']:>7.2f} | Rating: {p['rating']} | Title: {p['title']}")
else:
    print(f"Failed to fetch products. Status Code: {response.status_code}")

# --- Step 3: Search by Category ---
print("\n--- Laptops Category ---")
laptop_url = f"{base_url}/category/laptops"
laptop_res = requests.get(laptop_url)

if laptop_res.status_code == 200:
    laptops = laptop_res.json()["products"]
    for lp in laptops:
        print(f"Laptop: {lp['title']:<25} | Price: ${lp['price']:.2f}")

# --- Step 4: POST Request (Simulated) ---
print("\n--- Adding Custom Product (POST) ---")
add_url = f"{base_url}/add"
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

# Sending the POST request with json data
post_res = requests.post(add_url, json=new_product)

if post_res.status_code == 201 or post_res.status_code == 200:
    print("Success! Server Response:")
    print(post_res.json())
else:
    print(f"POST failed. Status: {post_res.status_code}")
```

```python
##Task 3 — Exception Handling
import requests

# --- Part A: Guarded Calculator ---
# This function handles the math and catches common input mistakes
def safe_divide(num1, num2):
    try:
        return num1 / num2
    except ZeroDivisionError:
        # User tried to divide by 0
        return "Error: Cannot divide by zero"
    except TypeError:
        # User passed a string or something that isn't a number
        return "Error: Invalid input types"

# Quick tests for the math function
print("--- Part A: Testing Calculator ---")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# --- Part B: Guarded File Reader ---
# Reading a file while ensuring we always report when the attempt finishes
def read_file_safe(target_name):
    try:
        with open(target_name, 'r', encoding="utf-8") as f:
            # Just return the whole text if it works
            return f.read()
    except FileNotFoundError:
        # File doesn't exist in the current directory
        print(f"Error: File '{target_name}' not found.")
        return ""
    finally:
        # The finally block is great for logs because it runs no matter what
        print("File operation attempt complete.")

print("\n--- Part B: Testing File Reader ---")
# Testing with the file from Task 1 (should work)
print(read_file_safe("python_notes.txt")[:60] + "...") 
# Testing a file that doesn't exist (should fail gracefully)
read_file_safe("ghost_file.txt")


# --- Part C & D: Product Lookup Loop ---
print("\n--- Part C & D: Product ID Lookup Tool ---")

while True:
    # Ask for input and clean it up (lowercase and no extra spaces)
    user_choice = input("\nEnter a product ID (1-100) or 'quit' to exit: ").strip().lower()

    if user_choice == 'quit':
        print("Closing the tool. Goodbye!")
        break

    # First, validate the input before even trying to use the API
    if not user_choice.isdigit():
        print("Warning: Please enter a whole number, not text.")
        continue

    item_id = int(user_choice)

    # Check if the number is within the dummyjson 1-100 range
    if item_id < 1 or item_id > 100:
        print("Warning: Please stay between 1 and 100.")
        continue

    # Now we try the network request
    try:
        url = f"https://dummyjson.com/products/{item_id}"
        # Set a 5-second timeout so the app doesn't hang on bad internet
        resp = requests.get(url, timeout=5)

        # Check for 404 (Resource not found)
        if resp.status_code == 404:
            print("Product not found.")
        elif resp.status_code == 200:
            # Success! Parse the JSON and show the title/price
            product_data = resp.json()
            print(f"Found: {product_data['title']} - ${product_data['price']}")
        else:
            print(f"Server error: Status {resp.status_code}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Check your internet connection.")
    except requests.exceptions.Timeout:
        print("The request timed out. The server might be slow.")
    except Exception as e:
        # General catch-all for any other weird errors
        print(f"Unexpected error: {e}")

print("\nAll tasks finished.")
```

```python
#Task 4 — Logging to File 

import requests
from datetime import datetime

# Simple helper to write errors to our log file
def log_error(location, error_type, message):
    # Get the current time in the exact format the task asks for
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # We use 'a' (append) so we don't delete old logs every time we run the script
    with open("error_log.txt", "a", encoding="utf-8") as f:
        f.write(f"[{timestamp}] ERROR in {location}: {error_type} — {message}\n")

print("--- Starting Task 4: Error Logging Test ---")

# 1. Triggering a ConnectionError (unreachable URL)
print("Attempting to connect to a fake website...")
try:
    # This URL definitely doesn't exist
    fake_url = "https://this-host-does-not-exist-xyz.com/api"
    requests.get(fake_url, timeout=5)
except requests.exceptions.ConnectionError as e:
    # Log the network failure
    log_error("fetch_products", "ConnectionError", "No connection could be made")
    print("✓ Recorded ConnectionError to log.")
except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e))


# 2. Triggering an HTTP 404 Error (Product 999)
print("Requesting a non-existent product ID (999)...")
try:
    bad_id_url = "https://dummyjson.com/products/999"
    resp = requests.get(bad_id_url, timeout=5)
    
    # Remember: 404 is a successful 'connection' but a failed 'find'
    if resp.status_code != 200:
        msg = f"{resp.status_code} Not Found for product ID 999"
        log_error("lookup_product", "HTTPError", msg)
        print("✓ Recorded 404 HTTPError to log.")
    else:
        print("Wait, product 999 actually exists?")
        
except Exception as e:
    log_error("lookup_product", "Exception", str(e))


# 3. Final Step: Read back the logs to prove it worked
print("\n--- Current Contents of error_log.txt ---")
try:
    with open("error_log.txt", "r", encoding="utf-8") as log_file:
        print(log_file.read())
except FileNotFoundError:
    print("The log file hasn't been created yet.")

print("Task 4 complete.")
```
